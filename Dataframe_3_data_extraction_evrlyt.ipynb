```python
#APPLY
def convert_to_str_and_add_mill(num):
  return str(num) + " MILL"
```

```python
bond["Box Office"] = bond["Box Office"].apply(convert_to_str_and_add_mill)
```

```python
bond.head(3)
def good_movie(row):
  actor=row[1]
  budget=row[4]
  
  if actor == 'Pierce Brosnan':
    return "BEST"
  elif actor == "Roger Moore" and budget >40:
    return "Enjoyable"
  else:
    return "NO IDEA"

bond.apply(good_movie,axis=1)
```
