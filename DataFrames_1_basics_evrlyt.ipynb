fillna

```python
# nba.fillna(0) #all nan in DF is replaced by 0
nba = players.copy()
nba.tail()
# nba["College"].fillna("Unknown",inplace=True) #inplace is not recc
nba["College"] = nba["College"].fillna("Unknown")
nba.tail()
```

```python
#ASTYPE
nba = pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/nba.csv").dropna(how="all")
nba.tail()
```

```python
nba["Age"].hasnans
nba["Age"].astype("int")
nba["Age"] = nba["Age"].astype("int")
nba["Salary"].hasnans
nba["Salary"]=nba["Salary"].fillna("0")
nba["Salary"].astype("int")
```

```python
nba["Position"].nunique()
nba.info()
```

```python
nba["Position"]=nba["Position"].astype('category')
nba.info() #RAM friendly
```

```python
# nba["Name"].sort_values()
nba.sort_values(by="Name",ascending=False)
nba.sort_values("Age",ascending=False)
nba.sort_values(by="Salary",na_position="first")
```

```python
nba = pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/nba.csv").dropna(how="all")
nba.tail()
```

```python
nba.sort_values(by=['Team','Name'])
nba.sort_values(by=['Name','Team'],ascending=False)
nba.sort_values(by=['Name','Team'],ascending=[False, True])
```

```python
nba.sort_index()
nba.sort_index(ascending=False)
```

```python
nba["Salary"]=nba["Salary"].fillna(0).astype('int')
nba["Salary Rank"]=nba["Salary"].rank(ascending=False)
nba.sort_values("Salary",ascending=False)
```
