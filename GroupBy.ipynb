```python
import pandas as pd
fortune=pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/fortune1000.csv",index_col="Rank")
sectors=fortune.groupby('Sector')
fortune.head()
```

```python
sectors = fortune.groupby(['Sector','Industry'])
sectors.size()
```

agg


```python
sectors = fortune.groupby('Sector')
```

```python
sectors.agg({'Revenue':'sum','Profits':'sum','Employees':'mean'})
```

```python
df = pd.DataFrame(columns=fortune.columns)
for sector, data in sectors:
  highest_revenue_company_in_group = data.nlargest(1,"Revenue")
  df=df.append(highest_revenue_company_in_group)

df
```

```python
cities = fortune.groupby("Location")
df=pd.DataFrame(columns=fortune.columns)
df
```

```python
for city,data in cities:
  high = data.nlargest(1,"Revenue")
  df=df.append(high)
df
```
