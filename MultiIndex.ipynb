```python
world=pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/worldstats.csv",index_col=["country","year"])
world.head()
```

```python
a.unstack()
a.unstack().unstack()
a.unstack().unstack().unstack()
```

pivot()

```python
sales=pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/salesmen.csv",parse_dates=["Date"])
sales['Salesman']=sales['Salesman'].astype('category')
sales.head()
```

```python
sales.pivot(index="Date",columns="Salesman",values="Revenue")
```

pivot_table()

```python
foods=pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/foods.csv")
foods.head(20)
```

```python
foods.pivot_table(values="Spend",index=["Item","Gender"],columns=["Frequency","City"],aggfunc='sum')
```

```python
foods.pivot_table(values="Spend",index=["Item","Gender"],columns="Frequency",aggfunc='count')
```

melt()

```python
sales=pd.read_csv("https://raw.githubusercontent.com/samarth-mistry/csv-data/main/pandas/quarters.csv")
sales.head()
```

```python
pd.melt(sales,id_vars="Salesman")
pd.melt(sales,id_vars="Salesman",var_name="Quarter",value_name="Revenue")
```
